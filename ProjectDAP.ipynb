```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
monthly_sales = df.groupby('month')['amount'].sum()
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='b')
plt.title("Total Sales Over Time")
plt.xlabel("Year-Month")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_categories = df.groupby("category")["amount"].sum().sort_values(ascending=False).head(10)
plt.bar(top_categories.index, top_categories.values, color='skyblue')
plt.title("Top 10 Best-Selling Product Categories")
plt.xlabel("Category")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(7, 7))
gender_sales = df.groupby("gender")["amount"].sum()
plt.pie(gender_sales, labels=gender_sales.index, autopct='%1.1f%%', colors=['#FF9999', '#66B3FF'], startangle=140)
plt.title("Sales Distribution by Gender")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_brands = df.groupby("brand")["amount"].sum().sort_values(ascending=False).head(10)
plt.bar(top_brands.index, top_brands.values, color='green')
plt.title("Top 10 Best-Selling Brands")
plt.xlabel("Brand")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["rating"], bins=5, color='purple', edgecolor='black', alpha=0.7)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
